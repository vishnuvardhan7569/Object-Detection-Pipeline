```python
import yaml
from pathlib import Path

data_yaml_path = "/kaggle/input/taco-dataset-yolo-format/data.yaml"

with open(data_yaml_path, "r") as f:
    data = yaml.safe_load(f)

data["path"] = "/kaggle/input/taco-dataset-yolo-format"
data["train"] = "train/images"
data["val"] = "valid/images"
data["test"] = "test/images"

with open("data_kaggle.yaml", "w") as f:
    yaml.dump(data, f)

print("‚úÖ data_kaggle.yaml created")
```
